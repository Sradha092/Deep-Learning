```python
# Import Basic Packages
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import DeepLearning Packages
import tensorflow as tf
from tensorflow import keras
```

```python
from sklearn.datasets import load_boston
```

```python
X,y = load_boston(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X = pd.DataFrame(X)
X.head()
```

```python
y = pd.DataFrame(y)
```

```python
y.head()
```

```python
sc = StandardScaler()
sc_x = sc.fit_transform(X)
sc_x
```

```python
X = pd.DataFrame(sc_x)
```

```python
X.head()
```

```python
# Split the data into train and test dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.75,random_state=1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Deep Learning Concept - Neural Networks

# Multilayer Perceptron

## API
- Sequential
- Functional

```python
# Setup for sequential API for building DNN

DNN = tf.keras.models.Sequential()
```

```python
# Input Layer connected with first hidden layer
DNN.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
# Input Layer connected with second hidden layer
DNN.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
# Adding Output Layer
DNN.add(tf.keras.layers.Dense(units=1))
```

```python
# Apply Optimization Techniques
# GD,SGD,Mini-batch SGD,SGD with momentum,Adagrade,adadelta,RMSProp,Adam
# Apply the measurement

DNN.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Apply input dataset along with Backpropagation method

DNN.fit(X_train, y_train, batch_size=32, epochs=500)
```

```python
DNN.fit(X_train, y_train, batch_size=32, epochs=100)
```

```python
# Predict the test data with training models
y_pred = DNN.predict(X_test)
print(y_test)
print('####'*20)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred_train=DNN.predict(X_train)
```

```python
print(r2_score(y_train,y_pred_train))
print('####'*20)
print(r2_score(y_test,y_pred))
```
